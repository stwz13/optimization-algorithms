```
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Define the function
def f(x):
    """Function to minimize"""
    x1, x2 = x
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Analytical solution (minimum)
def analytical_minimum():
    """The function has a known analytical minimum"""
    # From the structure: (5-x1)² = 0 => x1 = 5
    # Then (x2 - x1²)² = 0 => x2 = x1² = 25
    return np.array([5.0, 25.0])

# Create grid for contour plot
x1_grid = np.linspace(-3, 8, 500)
x2_grid = np.linspace(-5, 35, 500)
X1, X2 = np.meshgrid(x1_grid, x2_grid)
Z = f([X1, X2])

# Different optimization methods
initial_points = [
    np.array([0.0, 0.0]),      # Start from origin
    np.array([-2.0, 10.0]),    # Start from left side
    np.array([3.0, 5.0]),      # Start from near the valley
]

optimization_results = {}
methods = ['Nelder-Mead', 'BFGS', 'CG']

for method in methods:
    for i, x0 in enumerate(initial_points):
        result = minimize(f, x0, method=method)
        key = f"{method}_{i}"
        optimization_results[key] = {
            'x': result.x,
            'fun': result.fun,
            'success': result.success,
            'nit': result.nit,
            'x0': x0
        }

# Find the exact analytical minimum
exact_min = analytical_minimum()

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# First subplot: Level curves
ax1 = axes[0]
contour_levels = [0.1, 0.5, 1, 2, 5, 10, 20, 50, 100, 200, 500, 1000, 2000, 5000]
contour = ax1.contour(X1, X2, Z, levels=contour_levels, cmap='viridis', alpha=0.7)
ax1.clabel(contour, inline=True, fontsize=8, fmt='%d')
ax1.scatter(exact_min[0], exact_min[1], color='red', s=150,
           marker='*', zorder=5, label=f'Exact minimum ({exact_min[0]}, {exact_min[1]})')

# Add optimization paths
colors = ['blue', 'green', 'orange']
for idx, (key, result) in enumerate(optimization_results.items()):
    if result['success']:
        ax1.scatter(result['x'][0], result['x'][1], color=colors[idx % len(colors)],
                   s=80, marker='o', edgecolors='black', linewidth=1.5,
                   label=f"{key}: ({result['x'][0]:.3f}, {result['x'][1]:.3f})")

ax1.set_xlabel('x₁')
ax1.set_ylabel('x₂')
ax1.set_title('Level Curves and Optimization Results')
ax1.grid(alpha=0.3, linestyle='--')
ax1.legend(loc='upper left', fontsize=8)
ax1.set_xlim(-3, 8)
ax1.set_ylim(-5, 35)

# Second subplot: Zoom near minimum
ax2 = axes[1]
zoom_x1 = np.linspace(3, 7, 300)
zoom_x2 = np.linspace(20, 30, 300)
ZX1, ZX2 = np.meshgrid(zoom_x1, zoom_x2)
ZZ = f([ZX1, ZX2])

zoom_levels = [0.1, 0.5, 1, 2, 5, 10, 20, 50, 100]
contour_zoom = ax2.contour(ZX1, ZX2, ZZ, levels=zoom_levels, cmap='viridis')
ax2.clabel(contour_zoom, inline=True, fontsize=8)
ax2.scatter(exact_min[0], exact_min[1], color='red', s=150, marker='*', zorder=5)

for idx, (key, result) in enumerate(optimization_results.items()):
    if result['success'] and 3 <= result['x'][0] <= 7 and 20 <= result['x'][1] <= 30:
        ax2.scatter(result['x'][0], result['x'][1], color=colors[idx % len(colors)],
                   s=80, marker='o', edgecolors='black', linewidth=1.5)

ax2.set_xlabel('x₁')
ax2.set_ylabel('x₂')
ax2.set_title('Zoom Near Minimum (x₁∈[3,7], x₂∈[20,30])')
ax2.grid(alpha=0.3, linestyle='--')
ax2.set_xlim(3, 7)
ax2.set_ylim(20, 30)

plt.tight_layout()
plt.show()

# Print optimization results
print("=" * 60)
print("OPTIMIZATION RESULTS")
print("=" * 60)
print(f"\nExact analytical minimum: x₁ = {exact_min[0]}, x₂ = {exact_min[1]}, f = {f(exact_min):.10f}\n")

print("Numerical optimization results:")
print("-" * 60)
print(f"{'Method':<15} {'Start Point':<20} {'Found Minimum':<25} {'f(x)':<12} {'Iterations'}")
print("-" * 60)

for key, result in optimization_results.items():
    method, idx = key.split('_')
    start = result['x0']
    found = result['x']
    fun_val = result['fun']
    iterations = result['nit'] if 'nit' in result else 'N/A'
    status = "✓" if result['success'] else "✗"

    print(f"{method:<15} ({start[0]:.1f}, {start[1]:.1f})  ({found[0]:.6f}, {found[1]:.6f})  {fun_val:.2e}  {iterations:<8} {status}")

print("=" * 60)

# Additional: Find minimum using gradient information (if needed)
def gradient(x):
    """Gradient of the function for gradient-based methods"""
    x1, x2 = x
    df_dx1 = -2*(5 - x1) - 40*x1*(x2 - x1**2)
    df_dx2 = 20*(x2 - x1**2)
    return np.array([df_dx1, df_dx2])

# Try with gradient
result_gradient = minimize(f, [0, 0], method='BFGS', jac=gradient)
print(f"\nWith gradient (BFGS): x₁ = {result_gradient.x[0]:.6f}, x₂ = {result_gradient.x[1]:.6f}, f = {result_gradient.fun:.2e}")
```

```
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Точки, полученные в результате покоординатного спуска
# Итерация 1: из (0,0) -> (0.6, 0) -> (0.6, 0.36)
# Итерация 2: из (0.6, 0.36) -> (0.79, 0.36) -> (0.79, 0.52)
# Итерация 3: продолжаем процесс
points = [
    (0, 0),      # x1 начальная
    (0.6, 0),    # после шага по x1 на итерации 1
    (0.6, 0.36), # x2 после итерации 1
    (0.79, 0.36), # после шага по x1 на итерации 2
    (0.79, 0.52), # x3 после итерации 2
]

# Добавим еще несколько итераций для наглядности
# Итерация 3
# (0.79, 0.52) -> по x1: alpha ≈ 0.093 -> (0.883, 0.52)
# -> по x2: alpha = 0.883^2 = 0.78 -> (0.883, 0.78)
points.extend([(0.883, 0.52), (0.883, 0.78)])

# Итерация 4
# (0.883, 0.78) -> по x1: alpha ≈ 0.047 -> (0.93, 0.78)
# -> по x2: alpha = 0.93^2 = 0.865 -> (0.93, 0.865)
points.extend([(0.93, 0.78), (0.93, 0.865)])

# Итерация 5
# (0.93, 0.865) -> по x1: alpha ≈ 0.024 -> (0.954, 0.865)
# -> по x2: alpha = 0.954^2 = 0.91 -> (0.954, 0.91)
points.extend([(0.954, 0.865), (0.954, 0.91)])

# Преобразуем в массивы для удобства
points = np.array(points)

# Создаем сетку для линий уровня
x1 = np.linspace(-1, 6, 500)
x2 = np.linspace(-2, 12, 500)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Уровни для контуров
levels = [1, 5, 10, 20, 50, 100, 200, 500, 1000]

# Создаем фигуру с несколькими подграфиками
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# 1. Начальная точка
ax = axes[0]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начальная точка')
ax.plot(points[0, 0], points[0, 1], 'r*', markersize=12)
ax.set_title('Итерация 0: Начальная точка (0, 0)')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

# 2. После итерации 1 (до x2)
ax = axes[1]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:3, 0], points[:3, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[2, 0], points[2, 1], 'go', markersize=8, label='x² = (0.6, 0.36)')
ax.set_title('Итерация 1: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

# 3. После итерации 2
ax = axes[2]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:5, 0], points[:5, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[4, 0], points[4, 1], 'go', markersize=8, label='x³ = (0.79, 0.52)')
ax.set_title('Итерация 2: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

# 4. После итерации 3
ax = axes[3]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:7, 0], points[:7, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[6, 0], points[6, 1], 'go', markersize=8, label='x⁴ = (0.883, 0.78)')
ax.set_title('Итерация 3: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

# 5. После итерации 4
ax = axes[4]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:9, 0], points[:9, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[8, 0], points[8, 1], 'go', markersize=8, label='x⁵ = (0.93, 0.865)')
ax.set_title('Итерация 4: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

# 6. Полная траектория
ax = axes[5]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')

# Рисуем полную траекторию
for i in range(0, len(points)-1, 2):
    # Шаг по x1
    ax.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
            'r-', linewidth=2, alpha=0.7)
    if i+2 < len(points):
        # Шаг по x2
        ax.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                'b-', linewidth=2, alpha=0.7)

ax.plot(points[:, 0], points[:, 1], 'ko', markersize=4)
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начало (0, 0)')
ax.plot(points[-1, 0], points[-1, 1], 'g*', markersize=12, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
ax.plot(5, 25, 'r*', markersize=10, alpha=0.5, label='Истинный минимум (5, 25)')

ax.set_title('Полная траектория покоординатного спуска')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend(loc='upper left', fontsize=8)
ax.set_xlim(-1, 6)
ax.set_ylim(-2, 12)

plt.tight_layout()
plt.show()

# Отдельный большой график с полной траекторией
plt.figure(figsize=(12, 10))
contour = plt.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
plt.clabel(contour, inline=True, fontsize=8, fmt='%d')

# Рисуем траекторию с указанием направления
for i in range(0, len(points)-1, 2):
    # Шаг по x1 (красный)
    plt.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
             'r-', linewidth=2.5, alpha=0.8, label='Шаг по x₁' if i == 0 else "")
    # Добавляем стрелку
    mid_x = (points[i, 0] + points[i+1, 0]) / 2
    mid_y = (points[i, 1] + points[i+1, 1]) / 2
    dx = points[i+1, 0] - points[i, 0]
    dy = points[i+1, 1] - points[i, 1]
    plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
              head_width=0.08, head_length=0.08, fc='red', ec='red', alpha=0.8)

    if i+2 < len(points):
        # Шаг по x2 (синий)
        plt.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                 'b-', linewidth=2.5, alpha=0.8, label='Шаг по x₂' if i == 0 else "")
        # Добавляем стрелку
        mid_x = (points[i+1, 0] + points[i+2, 0]) / 2
        mid_y = (points[i+1, 1] + points[i+2, 1]) / 2
        dx = points[i+2, 0] - points[i+1, 0]
        dy = points[i+2, 1] - points[i+1, 1]
        plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
                  head_width=0.08, head_length=0.08, fc='blue', ec='blue', alpha=0.8)

plt.plot(points[:, 0], points[:, 1], 'ko', markersize=5)
plt.plot(points[0, 0], points[0, 1], 'ro', markersize=12, label='Начало (0, 0)')
plt.plot(points[-1, 0], points[-1, 1], 'g*', markersize=15, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
plt.plot(5, 25, 'r*', markersize=12, alpha=0.6, label='Истинный минимум (5, 25)')

plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('Метод покоординатного спуска: траектория движения к минимуму')
plt.grid(alpha=0.3, linestyle='--')
plt.legend(loc='upper left', fontsize=10)
plt.xlim(-1, 6)
plt.ylim(-2, 12)

plt.tight_layout()
plt.show()

# Вывод информации о точках
print("=" * 70)
print("Траектория покоординатного спуска:")
print("=" * 70)
print(f"{'Итерация':<10} {'Шаг':<10} {'x₁':<12} {'x₂':<12} {'f(x₁, x₂)':<15}")
print("-" * 70)

for i, (x, y) in enumerate(points):
    iter_num = i // 2 + 1 if i % 2 == 0 and i > 0 else i // 2 + 1 if i > 0 else 0
    step_type = "начало" if i == 0 else "по x₁" if i % 2 == 1 else "по x₂"
    if i == 0:
        print(f"{'0':<10} {step_type:<10} {x:<12.4f} {y:<12.4f} {f([x, y]):<15.4f}")
    else:
        print(f"{iter_num:<10} {step_type:<10} {x:<12.4f} {y:<12.4f} {f([x, y]):<15.4f}")

print("=" * 70)
print(f"Истинный минимум: x₁ = 5, x₂ = 25, f = 0")
print("=" * 70)
```

```
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Точки, полученные в результате покоординатного спуска
points = [
    (0, 0),      # x1 начальная
    (0.6, 0),    # после шага по x1 на итерации 1
    (0.6, 0.36), # x2 после итерации 1
    (0.79, 0.36), # после шага по x1 на итерации 2
    (0.79, 0.52), # x3 после итерации 2
]

points = np.array(points)

# Создаем сетку для линий уровня с увеличенным масштабом
x1 = np.linspace(0, 3, 500)
x2 = np.linspace(0, 3, 500)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Уровни для контуров
levels = [1, 5, 10, 20, 50, 100, 200, 500, 1000]

# Создаем фигуру с несколькими подграфиками
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# 1. Начальная точка
ax = axes[0]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начальная точка')
ax.plot(points[0, 0], points[0, 1], 'r*', markersize=12)
ax.set_title('Итерация 0: Начальная точка (0, 0)')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 2. После итерации 1 (до x2)
ax = axes[1]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:3, 0], points[:3, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[2, 0], points[2, 1], 'go', markersize=8, label='x² = (0.6, 0.36)')
ax.set_title('Итерация 1: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 3. После итерации 2
ax = axes[2]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:5, 0], points[:5, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[4, 0], points[4, 1], 'go', markersize=8, label='x³ = (0.79, 0.52)')
ax.set_title('Итерация 2: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 4. После итерации 3
ax = axes[3]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:7, 0], points[:7, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[6, 0], points[6, 1], 'go', markersize=8, label='x⁴ = (0.883, 0.78)')
ax.set_title('Итерация 3: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 5. После итерации 4
ax = axes[4]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:9, 0], points[:9, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[8, 0], points[8, 1], 'go', markersize=8, label='x⁵ = (0.93, 0.865)')
ax.set_title('Итерация 4: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 6. Полная траектория
ax = axes[5]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')

# Рисуем полную траекторию
for i in range(0, len(points)-1, 2):
    # Шаг по x1
    ax.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
            'r-', linewidth=2, alpha=0.7)
    if i+2 < len(points):
        # Шаг по x2
        ax.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                'b-', linewidth=2, alpha=0.7)

ax.plot(points[:, 0], points[:, 1], 'ko', markersize=4)
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начало (0, 0)')
ax.plot(points[-1, 0], points[-1, 1], 'g*', markersize=12, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
ax.plot(5, 25, 'r*', markersize=10, alpha=0.5, label='Истинный минимум (5, 25)')

ax.set_title('Полная траектория покоординатного спуска')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend(loc='upper left', fontsize=8)
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

plt.tight_layout()
plt.show()

# Отдельный большой график с полной траекторией (увеличенный масштаб)
plt.figure(figsize=(12, 10))
contour = plt.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
plt.clabel(contour, inline=True, fontsize=8, fmt='%d')

# Рисуем траекторию с указанием направления
for i in range(0, len(points)-1, 2):
    # Шаг по x1 (красный)
    plt.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
             'r-', linewidth=2.5, alpha=0.8, label='Шаг по x₁' if i == 0 else "")
    # Добавляем стрелку
    mid_x = (points[i, 0] + points[i+1, 0]) / 2
    mid_y = (points[i, 1] + points[i+1, 1]) / 2
    dx = points[i+1, 0] - points[i, 0]
    dy = points[i+1, 1] - points[i, 1]
    plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
              head_width=0.03, head_length=0.03, fc='red', ec='red', alpha=0.8)

    if i+2 < len(points):
        # Шаг по x2 (синий)
        plt.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                 'b-', linewidth=2.5, alpha=0.8, label='Шаг по x₂' if i == 0 else "")
        # Добавляем стрелку
        mid_x = (points[i+1, 0] + points[i+2, 0]) / 2
        mid_y = (points[i+1, 1] + points[i+2, 1]) / 2
        dx = points[i+2, 0] - points[i+1, 0]
        dy = points[i+2, 1] - points[i+1, 1]
        plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
                  head_width=0.03, head_length=0.03, fc='blue', ec='blue', alpha=0.8)

plt.plot(points[:, 0], points[:, 1], 'ko', markersize=5)
plt.plot(points[0, 0], points[0, 1], 'ro', markersize=12, label='Начало (0, 0)')
plt.plot(points[-1, 0], points[-1, 1], 'g*', markersize=15, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
plt.plot(5, 25, 'r*', markersize=12, alpha=0.4, label='Истинный минимум (5, 25)')

plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('Метод покоординатного спуска: траектория движения к минимуму (масштаб 0-3)')
plt.grid(alpha=0.3, linestyle='--')
plt.legend(loc='upper left', fontsize=10)
plt.xlim(0, 3)
plt.ylim(0, 3)

plt.tight_layout()
plt.show()

# Вывод информации о точках
print("=" * 70)
print("Траектория покоординатного спуска:")
print("=" * 70)
print(f"{'Итерация':<10} {'Шаг':<10} {'x₁':<12} {'x₂':<12} {'f(x₁, x₂)':<15}")
print("-" * 70)

for i, (x, y) in enumerate(points):
    if i == 0:
        iter_num = 0
        step_type = "начало"
    elif i % 2 == 1:
        iter_num = (i // 2) + 1
        step_type = "по x₁"
    else:
        iter_num = i // 2
        step_type = "по x₂"

    print(f"{iter_num:<10} {step_type:<10} {x:<12.4f} {y:<12.4f} {f([x, y]):<15.4f}")

print("=" * 70)
print(f"Истинный минимум: x₁ = 5, x₂ = 25, f = 0")
print("=" * 70)
```

```
import numpy as np
import matplotlib.pyplot as plt

# 1. Определение функции и данных
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Координаты точек, рассчитанные в ходе итераций (x, y)
# Каждая итерация состоит из двух шагов: по x1 и по x2
path = [
    (0, 0),         # Начало (Итерация 0)
    (0.6, 0),       # Ит 1: шаг по x1
    (0.6, 0.36),    # Ит 1: шаг по x2 (x^2)
    (0.79, 0.36),   # Ит 2: шаг по x1
    (0.79, 0.62),   # Ит 2: шаг по x2 (x^3)
    (0.88, 0.62),   # Ит 3: шаг по x1
    (0.88, 0.78),   # Ит 3: шаг по x2 (x^4)
    (0.93, 0.78),   # Ит 4: шаг по x1
    (0.93, 0.86)    # Ит 4: шаг по x2 (x^5)
]

# Настройки сетки для изолиний
x1_vals = np.linspace(0, 3, 400)
x2_vals = np.linspace(0, 3, 400)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
Z = f(X1, X2)
levels = [1, 5, 10, 20, 50, 100, 200, 500]

# 2. Создание полотна с 6 подграфиками
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# Заголовки для каждого окна
titles = [
    "Итерация 0: Начальная точка",
    "Итерация 1: После шагов по x1 и x2",
    "Итерация 2: После шагов по x1 и x2",
    "Итерация 3: После шагов по x1 и x2",
    "Итерация 4: После шагов по x1 и x2",
    "Полная траектория спуска"
]

# 3. Цикл отрисовки
for i in range(6):
    ax = axes[i]
    # Рисуем линии уровня
    cp = ax.contour(X1, X2, Z, levels=levels, colors='gray', alpha=0.5, linewidths=1)
    ax.clabel(cp, inline=True, fontsize=8)

    if i == 0:
        # Только начальная точка
        ax.scatter(path[0][0], path[0][1], color='red', s=100, label='Начальная точка')
    elif i < 5:
        # Отрисовка пути до текущей итерации (2*i шагов)
        current_path = np.array(path[:2*i + 1])
        ax.plot(current_path[:, 0], current_path[:, 1], 'b-o', markersize=4, label='Траектория')
        # Последняя точка итерации
        ax.scatter(current_path[-1, 0], current_path[-1, 1], color='green', s=60,
                   zorder=5, label=f'x^{i+1} = ({current_path[-1, 0]}, {current_path[-1, 1]})')
    else:
        # Финальный график со всеми точками и истинным минимумом
        full_path = np.array(path)
        ax.plot(full_path[:, 0], full_path[:, 1], 'k-o', markersize=3, alpha=0.7)
        ax.scatter(path[0][0], path[0][1], color='red', label='Начало (0,0)')
        ax.scatter(full_path[-1, 0], full_path[-1, 1], color='green', label='Конец (0.93, 0.86)')
        # Истинный минимум функции (5, 25) вне текущего масштаба, но пометим направление
        ax.scatter(5, 25, color='orange', marker='*', s=200, label='Истинный минимум (5, 25)')

    # Оформление каждого графика
    ax.set_title(titles[i])
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.set_xlim(0, 3)
    ax.set_ylim(0, 3)
    ax.legend(loc='lower right', fontsize='small')

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# 1. Определение функции
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# 2. Список всех промежуточных точек (расчетные данные)
# Точки: x1, y1_2, x2, y2_2, x3, y3_2, x4, y4_2, x5
path = [
    (0, 0),         # x1
    (0.6, 0),       # y1_2
    (0.6, 0.36),    # x2
    (0.79, 0.36),   # y2_2
    (0.79, 0.62),   # x3
    (0.88, 0.62),   # y3_2
    (0.88, 0.78),   # x4
    (0.93, 0.78),   # y4_2
    (0.93, 0.86)    # x5
]

# Сетка для графиков
x1_vals = np.linspace(-0.2, 2.5, 500)
x2_vals = np.linspace(-0.2, 2.5, 500)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
Z = f(X1, X2)

# 3. Настройка фигуры
fig, axes = plt.subplots(2, 3, figsize=(18, 11))
axes = axes.flatten()

titles = [
    "Ит. 0: Старт", "Ит. 1: Шаги по x1, x2",
    "Ит. 2: Шаги по x1, x2", "Ит. 3: Шаги по x1, x2",
    "Ит. 4: Шаги по x1, x2", "Итоговая траектория"
]

for i in range(6):
    ax = axes[i]
    # Общий фон (светлые линии уровня)
    bg_levels = np.logspace(0.5, 3, 15)
    ax.contour(X1, X2, Z, levels=bg_levels, colors='lightgray', alpha=0.3)

    if i < 5:
        # Индекс последней точки для текущего кадра (2 шага на итерацию)
        idx_end = 2 * i
        current_points = np.array(path[:idx_end + 1])

        # Отрисовка пути
        ax.plot(current_points[:, 0], current_points[:, 1], 'b-o', markersize=4, alpha=0.6)

        # ВЫДЕЛЕНИЕ ЛИНИИ УРОВНЯ ДЛЯ ТЕКУЩЕЙ ТОЧКИ
        curr_x, curr_y = path[idx_end]
        val = f(curr_x, curr_y)
        # Рисуем жирную линию уровня, проходящую через текущую точку
        cs = ax.contour(X1, X2, Z, levels=[val], colors='red', linewidths=2)
        ax.clabel(cs, fmt={val: f'f={val:.2f}'}, inline=True, fontsize=10)

        ax.scatter(curr_x, curr_y, color='red', s=50, zorder=5)
        ax.set_title(f"{titles[i]}\nТекущая точка: ({curr_x}, {curr_y})")

    else:
        # Финальный график
        full_path = np.array(path)
        ax.plot(full_path[:, 0], full_path[:, 1], 'k-o', markersize=3)
        # Подсветка нескольких ключевых уровней
        for pt_idx in [0, 2, 4, 8]:
            pt = path[pt_idx]
            v = f(pt[0], pt[1])
            ax.contour(X1, X2, Z, levels=[v], colors='blue', alpha=0.5, linestyles='--')

        ax.set_title(titles[i])
        ax.scatter(5, 25, color='gold', marker='*', s=200, label='Цель (5, 25)')

    ax.set_xlim(-0.1, 2.0)
    ax.set_ylim(-0.1, 2.0)
    ax.grid(True, linestyle=':', alpha=0.6)

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# 1. Функция и расчетные точки
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Координаты из ваших расчетов: x1, y1.2, x2, y2.2, x3, y3.2, x4, y4.2, x5
path = [
    (0, 0),         # x1
    (0.6, 0),       # y1.2 (минимум по x1)
    (0.6, 0.36),    # x2   (минимум по x2)
    (0.79, 0.36),   # y2.2
    (0.79, 0.62),   # x3
    (0.88, 0.62),   # y3.2
    (0.88, 0.78),   # x4
    (0.93, 0.78),   # y4.2
    (0.93, 0.86)    # x5
]

# Сетка для построения
x1_v = np.linspace(-0.2, 2.0, 500)
x2_v = np.linspace(-0.2, 2.0, 500)
X1, X2 = np.meshgrid(x1_v, x2_v)
Z = f(X1, X2)

# Цвета для уровней (чтобы было наглядно)
colors = ['red', 'orange', 'gold', 'green', 'cyan', 'blue', 'purple', 'magenta', 'brown']

# 2. Создание графиков
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

titles = [
    "Итерация 0: Начало", "Итерация 1: Шаги по x1 и x2",
    "Итерация 2: Шаги по x1 и x2", "Итерация 3: Шаги по x1 и x2",
    "Итерация 4: Шаги по x1 и x2", "Полная траектория"
]

for i in range(6):
    ax = axes[i]

    # Определяем, какие точки показывать на текущем графике
    if i < 5:
        idx_limit = 2 * i + 1
    else:
        idx_limit = len(path)

    current_points = np.array(path[:idx_limit])

    # 3. Отрисовка ЛИНИЙ УРОВНЯ только для текущих точек
    unique_levels = []
    for j in range(idx_limit):
        p = path[j]
        val = f(p[0], p[1])
        # Чтобы не рисовать одну и ту же линию дважды
        if not any(np.isclose(val, unique_levels)):
            unique_levels.append(val)
            # Рисуем цветную линию уровня
            cs = ax.contour(X1, X2, Z, levels=[val], colors=colors[j % len(colors)], linewidths=1.5)
            ax.clabel(cs, fmt={val: f'{val:.2f}'}, inline=True, fontsize=9)

    # 4. Отрисовка траектории
    ax.plot(current_points[:, 0], current_points[:, 1], 'k-', alpha=0.3) # Линия пути
    ax.scatter(current_points[:, 0], current_points[:, 1], c='black', s=20, zorder=3) # Все точки черные

    # Выделяем последнюю точку итерации
    last_p = current_points[-1]
    ax.scatter(last_p[0], last_p[1], color='red', s=80, edgecolors='black', zorder=5, label='Текущая точка')

    # Оформление
    ax.set_title(titles[i], fontsize=14, fontweight='bold')
    ax.set_xlim(-0.1, 1.5)
    ax.set_ylim(-0.1, 1.5)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.grid(True, linestyle=':', alpha=0.4)
    if i == 0 or i == 5: ax.legend()

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# 1. Функция Розенброка (модифицированная)
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# 2. Рассчитанные точки пути (x1, x2)
path = [
    (0, 0),         # Начало
    (0.6, 0),       # Ит 1.1
    (0.6, 0.36),    # Ит 1.2 (точка x2)
    (0.79, 0.36),   # Ит 2.1
    (0.79, 0.62),   # Ит 2.2 (точка x3)
    (0.88, 0.62),   # Ит 3.1
    (0.88, 0.78),   # Ит 3.2 (точка x4)
    (0.93, 0.78),   # Ит 4.1
    (0.93, 0.86)    # Ит 4.2 (точка x5)
]

# Сетка для отрисовки
x_range = np.linspace(-0.2, 2.5, 500)
y_range = np.linspace(-0.2, 2.5, 500)
X1, X2 = np.meshgrid(x_range, y_range)
Z = f(X1, X2)

# Уровни для фоновых линий
bg_levels = np.logspace(0.5, 3, 15)

# 3. Настройка графиков
fig, axes = plt.subplots(2, 3, figsize=(18, 11))
axes = axes.flatten()

titles = [
    "Итерация 0: Старт", "Итерация 1",
    "Итерация 2", "Итерация 3",
    "Итерация 4", "Полная траектория"
]

for i in range(6):
    ax = axes[i]

    # А. Рисуем фоновые линии уровня (теперь светло-голубые)
    ax.contour(X1, X2, Z, levels=bg_levels, colors='lightblue', alpha=0.4, linewidths=0.8)

    if i < 5:
        idx_end = 2 * i
        current_path = np.array(path[:idx_end + 1])

        # Б. Выделяем ПУНКТИРОМ линию уровня текущей точки
        curr_p = path[idx_end]
        val = f(*curr_p)
        cs = ax.contour(X1, X2, Z, levels=[val], colors='darkorange', linestyles='--', linewidths=2)
        ax.clabel(cs, fmt={val: f'f={val:.2f}'}, inline=True, fontsize=9)

        # В. Отрисовка траектории
        ax.plot(current_path[:, 0], current_path[:, 1], 'b-o', markersize=4, linewidth=1.5, label='Путь')
        ax.scatter(*curr_p, color='red', s=60, zorder=5, edgecolors='black')

    else:
        # Г. Финальный график: все пунктирные линии пройденных этапов
        full_path = np.array(path)
        ax.plot(full_path[:, 0], full_path[:, 1], 'b-o', markersize=3, alpha=0.5)

        # Рисуем все ключевые уровни пунктиром
        for step in [0, 2, 4, 6, 8]:
            p_step = path[step]
            v_step = f(*p_step)
            ax.contour(X1, X2, Z, levels=[v_step], colors='darkorange', linestyles='--', alpha=0.6)

        ax.scatter(5, 25, color='gold', marker='*', s=200, label='Минимум (5, 25)')

    # Оформление
    ax.set_title(titles[i], fontsize=13)
    ax.set_xlim(-0.1, 1.8)
    ax.set_ylim(-0.1, 1.8)
    ax.set_xlabel('x1')
    ax.set_ylabel('x2')
    ax.grid(True, linestyle=':', alpha=0.3)
    if i == 5: ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Определяем функцию
def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Точки, полученные в результате покоординатного спуска
points = [
    (0, 0),      # x1 начальная
    (0.6, 0),    # после шага по x1 на итерации 1
    (0.6, 0.36), # x2 после итерации 1
    (0.79, 0.36), # после шага по x1 на итерации 2
    (0.79, 0.52), # x3 после итерации 2
    (0.883, 0.52), # после шага по x1 на итерации 3
    (0.883, 0.78), # x4 после итерации 3
]

points = np.array(points)

# Создаем сетку для линий уровня с увеличенным масштабом
x1 = np.linspace(0, 3, 500)
x2 = np.linspace(0, 3, 500)
X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)

# Уровни для контуров
levels = [1, 5, 10, 20, 50, 100, 200, 500, 1000]

# Создаем фигуру с несколькими подграфиками
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

# 1. Начальная точка
ax = axes[0]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начальная точка')
ax.plot(points[0, 0], points[0, 1], 'r*', markersize=12)
ax.set_title('Итерация 0: Начальная точка (0, 0)')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 2. После итерации 1 (до x2)
ax = axes[1]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:3, 0], points[:3, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[2, 0], points[2, 1], 'go', markersize=8, label='x² = (0.6, 0.36)')
ax.set_title('Итерация 1: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 3. После итерации 2
ax = axes[2]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:5, 0], points[:5, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[4, 0], points[4, 1], 'go', markersize=8, label='x³ = (0.79, 0.52)')
ax.set_title('Итерация 2: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 4. После итерации 3
ax = axes[3]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:7, 0], points[:7, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[6, 0], points[6, 1], 'go', markersize=8, label='x⁴ = (0.883, 0.78)')
ax.set_title('Итерация 3: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 5. После итерации 4
ax = axes[4]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')
ax.plot(points[:9, 0], points[:9, 1], 'b-o', markersize=6, linewidth=2, label='Траектория')
ax.plot(points[8, 0], points[8, 1], 'go', markersize=8, label='x⁵ = (0.93, 0.865)')
ax.set_title('Итерация 4: После шагов по x₁ и x₂')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend()
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# 6. Полная траектория
ax = axes[5]
contour = ax.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
ax.clabel(contour, inline=True, fontsize=7, fmt='%d')

# Рисуем полную траекторию
for i in range(0, len(points)-1, 2):
    # Шаг по x1
    ax.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
            'r-', linewidth=2, alpha=0.7)
    if i+2 < len(points):
        # Шаг по x2
        ax.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                'b-', linewidth=2, alpha=0.7)

ax.plot(points[:, 0], points[:, 1], 'ko', markersize=4)
ax.plot(points[0, 0], points[0, 1], 'ro', markersize=10, label='Начало (0, 0)')
ax.plot(points[-1, 0], points[-1, 1], 'g*', markersize=12, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
ax.plot(5, 25, 'r*', markersize=10, alpha=0.5, label='Истинный минимум (5, 25)')

ax.set_title('Полная траектория покоординатного спуска')
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.grid(alpha=0.3)
ax.legend(loc='upper left', fontsize=8)
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

plt.tight_layout()
plt.show()

# Отдельный большой график с полной траекторией (увеличенный масштаб)
plt.figure(figsize=(12, 10))
contour = plt.contour(X1, X2, Z, levels=levels, cmap='viridis', alpha=0.7)
plt.clabel(contour, inline=True, fontsize=8, fmt='%d')

# Рисуем траекторию с указанием направления
for i in range(0, len(points)-1, 2):
    # Шаг по x1 (красный)
    plt.plot([points[i, 0], points[i+1, 0]], [points[i, 1], points[i+1, 1]],
             'r-', linewidth=2.5, alpha=0.8, label='Шаг по x₁' if i == 0 else "")
    # Добавляем стрелку
    mid_x = (points[i, 0] + points[i+1, 0]) / 2
    mid_y = (points[i, 1] + points[i+1, 1]) / 2
    dx = points[i+1, 0] - points[i, 0]
    dy = points[i+1, 1] - points[i, 1]
    plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
              head_width=0.03, head_length=0.03, fc='red', ec='red', alpha=0.8)

    if i+2 < len(points):
        # Шаг по x2 (синий)
        plt.plot([points[i+1, 0], points[i+2, 0]], [points[i+1, 1], points[i+2, 1]],
                 'b-', linewidth=2.5, alpha=0.8, label='Шаг по x₂' if i == 0 else "")
        # Добавляем стрелку
        mid_x = (points[i+1, 0] + points[i+2, 0]) / 2
        mid_y = (points[i+1, 1] + points[i+2, 1]) / 2
        dx = points[i+2, 0] - points[i+1, 0]
        dy = points[i+2, 1] - points[i+1, 1]
        plt.arrow(mid_x - dx*0.1, mid_y - dy*0.1, dx*0.2, dy*0.2,
                  head_width=0.03, head_length=0.03, fc='blue', ec='blue', alpha=0.8)

plt.plot(points[:, 0], points[:, 1], 'ko', markersize=5)
plt.plot(points[0, 0], points[0, 1], 'ro', markersize=12, label='Начало (0, 0)')
plt.plot(points[-1, 0], points[-1, 1], 'g*', markersize=15, label=f'Конец ({points[-1,0]:.3f}, {points[-1,1]:.3f})')
plt.plot(5, 25, 'r*', markersize=12, alpha=0.4, label='Истинный минимум (5, 25)')

plt.xlabel('x₁')
plt.ylabel('x₂')
plt.title('Метод покоординатного спуска: траектория движения к минимуму (масштаб 0-3)')
plt.grid(alpha=0.3, linestyle='--')
plt.legend(loc='upper left', fontsize=10)
plt.xlim(0, 3)
plt.ylim(0, 3)

plt.tight_layout()
plt.show()

# Вывод информации о точках
print("=" * 70)
print("Траектория покоординатного спуска:")
print("=" * 70)
print(f"{'Итерация':<10} {'Шаг':<10} {'x₁':<12} {'x₂':<12} {'f(x₁, x₂)':<15}")
print("-" * 70)

for i, (x, y) in enumerate(points):
    if i == 0:
        iter_num = 0
        step_type = "начало"
    elif i % 2 == 1:
        iter_num = (i // 2) + 1
        step_type = "по x₁"
    else:
        iter_num = i // 2
        step_type = "по x₂"

    print(f"{iter_num:<10} {step_type:<10} {x:<12.4f} {y:<12.4f} {f([x, y]):<15.4f}")

print("=" * 70)
print(f"Истинный минимум: x₁ = 5, x₂ = 25, f = 0")
print("=" * 70)
```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Точки покоординатного спуска
points = [
    (0, 0), (0.6, 0), (0.6, 0.36), (0.79, 0.36),
    (0.79, 0.52), (0.883, 0.52), (0.883, 0.78)
]

x1_grid = np.linspace(-0.2, 1.2, 500)
x2_grid = np.linspace(-0.2, 1.2, 500)
X1, X2 = np.meshgrid(x1_grid, x2_grid)
Z = f(X1, X2)

fig, axes = plt.subplots(4, 2, figsize=(15, 22))
axes = axes.flatten()

# Базовые синие уровни
base_levels = np.linspace(0, 30, 12)

for i, point in enumerate(points):
    ax = axes[i]
    cur_x1, cur_x2 = point
    val = f(cur_x1, cur_x2)

    # 1. Синие линии уровня с подписями
    cs_blue = ax.contour(X1, X2, Z, levels=base_levels, colors='blue', alpha=0.2, linewidths=0.8)
    ax.clabel(cs_blue, inline=True, fontsize=8, fmt='%.1f')

    # 2. Текущая линия уровня (красный пунктир) с подписью
    cs_red = ax.contour(X1, X2, Z, levels=[val], colors='red', linestyles='dashed', linewidths=2)
    ax.clabel(cs_red, inline=True, fontsize=9, fmt={val: f'{val:.2f}'})

    # 3. Отрисовка траектории (серая линия)
    path = np.array(points[:i+1])
    ax.plot(path[:, 0], path[:, 1], 'k-', alpha=0.2, zorder=3)

    # 4. Отрисовка точек с выделением цветов для легенды
    legend_elements = []

    for j in range(i + 1):
        p_x, p_y = points[j]
        point_val = f(p_x, p_y)

        if j == i: # Текущая точка
            scatter = ax.scatter(p_x, p_y, color='red', s=60, zorder=5)
            legend_elements.append((scatter, f'({p_x}, {p_y}), f={point_val:.2f}'))
        elif j == i - 1: # Предыдущая точка
            scatter = ax.scatter(p_x, p_y, color='orange', s=50, zorder=4)
            legend_elements.append((scatter, f'({p_x}, {p_y}), f={point_val:.2f}'))
        else: # Остальные пройденные
            ax.scatter(p_x, p_y, color='black', s=20, alpha=0.3, zorder=3)

    # Добавляем красную линию уровня в легенду
    from matplotlib.lines import Line2D
    red_line = Line2D([0], [0], color='red', linestyle='dashed', linewidth=2, label=f'Линия уровня f={val:.2f}')
    legend_elements.append((red_line, red_line.get_label()))

    ax.set_title(f'Шаг {i}', fontsize=12)
    ax.set_xlim(-0.1, 1.1)
    ax.set_ylim(-0.1, 1.0)

    # Создаем легенду
    if legend_elements:
        handles, labels = zip(*legend_elements)
        ax.legend(handles, labels, loc='upper left', fontsize=10, framealpha=0.9)

# 8-й график: Итоговая траектория
ax_total = axes[-1]
ax_total.contour(X1, X2, Z, levels=base_levels, colors='blue', alpha=0.15)
all_path = np.array(points)
ax_total.plot(all_path[:, 0], all_path[:, 1], 'k-', alpha=0.3)

legend_elements_total = []
for idx, p in enumerate(points):
    v = f(p[0], p[1])
    cs_red_total = ax_total.contour(X1, X2, Z, levels=[v], colors='red', linestyles='dashed', alpha=0.4)
    ax_total.clabel(cs_red_total, inline=True, fontsize=7, fmt={v: f'{v:.2f}'})
    scatter = ax_total.scatter(p[0], p[1], s=30)
    legend_elements_total.append((scatter, f'({p[0]}, {p[1]}), f={v:.2f}'))

ax_total.set_title('Итоговая траектория', fontsize=12)
handles_total, labels_total = zip(*legend_elements_total)
ax_total.legend(handles_total, labels_total, loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8, title="Все координаты:")

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

# Точки для метода покоординатного спуска
coord_points = [
    (0, 0), (0.6, 0), (0.6, 0.36), (0.79, 0.36),
    (0.79, 0.52), (0.883, 0.52), (0.883, 0.78)
]

# Точки для метода Розенброка (из расчетов)
rosenbrock_points = [
    (0, 0),           # начальная точка
    (0.6, 0.36),      # после итерации 1
    (0.94, 0.7),      # после итерации 2
    (1.16, 1.11)      # после итерации 3
]

# Направления для метода Розенброка (для отображения)
rosenbrock_dirs = [
    [(0.6, 0.36), (-0.1588, 0.2647)],  # итерация 1
    [(0.34, 0.33), (-0.08, 0.08)],     # итерация 2
    [(0.23, 0.41), (-0.12, 0.07)]      # итерация 3
]

# Создаем сетку для линий уровня
x1_grid = np.linspace(-0.5, 2.0, 500)
x2_grid = np.linspace(-0.5, 1.5, 500)
X1, X2 = np.meshgrid(x1_grid, x2_grid)
Z = f(X1, X2)

fig, ax = plt.subplots(1, 1, figsize=(14, 10))

# Рисуем линии уровня
levels = np.array([0.5, 1, 2, 5, 10, 15, 20, 25, 30, 40, 50, 60, 80, 100])
cs = ax.contour(X1, X2, Z, levels=levels, colors='blue', alpha=0.4, linewidths=0.8)
ax.clabel(cs, inline=True, fontsize=8, fmt='%.0f')

# Более плотные линии уровня в области минимума
levels_dense = np.linspace(0.1, 1, 10)
cs_dense = ax.contour(X1, X2, Z, levels=levels_dense, colors='lightblue', alpha=0.3, linewidths=0.5, linestyles='dotted')

# Метод покоординатного спуска
coord_path = np.array(coord_points)
ax.plot(coord_path[:, 0], coord_path[:, 1], 'g-', linewidth=2, alpha=0.7, label='Покоординатный спуск')
ax.scatter(coord_path[:, 0], coord_path[:, 1], color='green', s=40, zorder=5)

# Подписываем точки покоординатного спуска
for i, (x, y) in enumerate(coord_points):
    if i == 0:
        ax.annotate(f'Начало\n({x}, {y})', (x, y), xytext=(x-0.15, y-0.1), fontsize=8,
                   bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))
    elif i == len(coord_points)-1:
        ax.annotate(f'Конец\n({x:.3f}, {y:.3f})', (x, y), xytext=(x+0.05, y+0.05), fontsize=8,
                   bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.7))
    else:
        ax.annotate(f'{i}\n({x:.2f}, {y:.2f})', (x, y), xytext=(x+0.02, y+0.02), fontsize=7,
                   bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.6))

# Метод Розенброка
rosen_path = np.array(rosenbrock_points)
ax.plot(rosen_path[:, 0], rosen_path[:, 1], 'r-', linewidth=2, alpha=0.7, label='Метод Розенброка')
ax.scatter(rosen_path[:, 0], rosen_path[:, 1], color='red', s=50, zorder=5, marker='s')

# Подписываем точки Розенброка
for i, (x, y) in enumerate(rosenbrock_points):
    if i == 0:
        ax.annotate(f'({x}, {y})', (x, y), xytext=(x-0.15, y-0.12), fontsize=8, color='red',
                   bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.7))
    else:
        ax.annotate(f'Ит.{i}\n({x:.2f}, {y:.2f})', (x, y), xytext=(x+0.05, y+0.05), fontsize=8, color='red',
                   bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.7))

# Рисуем направления поиска для метода Розенброка
colors_dir = ['orange', 'purple', 'brown']
for iter_idx, (start_point, directions) in enumerate(zip(rosenbrock_points[:-1], rosenbrock_dirs)):
    start = np.array(start_point)
    for dir_idx, direction in enumerate(directions):
        dir_vec = np.array(direction)
        # Находим следующую точку после движения по этому направлению
        if iter_idx == 0:
            if dir_idx == 0:
                end_point = rosenbrock_points[1]  # после первого направления
            else:
                # Это промежуточная точка, которая не входит в основной список
                # Вычисляем её из расчетов
                if dir_idx == 1:
                    end_point = (0.996, 0.5976)  # y^2 после первого направления
        elif iter_idx == 1:
            if dir_idx == 0:
                end_point = (1.263, 1.0135)
            else:
                end_point = (1.16, 1.11)
        else:
            if dir_idx == 0:
                end_point = (1.263, 1.0135)
            else:
                end_point = (1.16, 1.11)

        # Рисуем стрелку направления
        ax.arrow(start[0], start[1], dir_vec[0]*0.5, dir_vec[1]*0.5,
                head_width=0.03, head_length=0.05, fc=colors_dir[iter_idx], ec=colors_dir[iter_idx],
                alpha=0.5, linestyle='--', linewidth=1)

# Отмечаем точку минимума (приблизительно)
x_min, y_min = 1.0, 1.0  # приблизительный минимум для Rosenbrock функции
ax.scatter(x_min, y_min, color='gold', s=100, zorder=6, marker='*', edgecolors='black', label='Область минимума')

# Настройка графика
ax.set_xlabel('$x_1$', fontsize=12)
ax.set_ylabel('$x_2$', fontsize=12)
ax.set_title('Сравнение метода покоординатного спуска и метода Розенброка\nна функции Розенброка $f(x_1, x_2) = (5-x_1)^2 + 10(x_2 - x_1^2)^2$',
             fontsize=14, pad=20)
ax.set_xlim(-0.2, 1.8)
ax.set_ylim(-0.2, 1.3)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

# Легенда
legend_elements = [
    Line2D([0], [0], color='green', linewidth=2, label='Покоординатный спуск'),
    Line2D([0], [0], color='red', linewidth=2, label='Метод Розенброка'),
    Line2D([0], [0], color='blue', linewidth=0.8, label='Линии уровня'),
    Line2D([0], [0], marker='s', color='red', linestyle='', markersize=8, label='Точки Розенброка'),
    Line2D([0], [0], marker='o', color='green', linestyle='', markersize=6, label='Точки покоординатного спуска'),
    Line2D([0], [0], marker='*', color='gold', linestyle='', markersize=12, label='Область минимума'),
]

ax.legend(handles=legend_elements, loc='upper right', fontsize=10, framealpha=0.9)

# Добавляем текстовую информацию
text_info = f"""Метод покоординатного спуска:
• Начальная точка: (0, 0)
• Конечная точка: ({coord_points[-1][0]:.3f}, {coord_points[-1][1]:.3f})
• Значение функции: {f(*coord_points[-1]):.2f}

Метод Розенброка:
• Начальная точка: (0, 0)
• Конечная точка: ({rosenbrock_points[-1][0]:.2f}, {rosenbrock_points[-1][1]:.2f})
• Значение функции: {f(*rosenbrock_points[-1]):.2f}"""

ax.text(1.05, 0.02, text_info, transform=ax.transAxes, fontsize=9,
        verticalalignment='bottom', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.show()

# Дополнительный график: только метод Розенброка с направлениями поиска
fig2, ax2 = plt.subplots(1, 1, figsize=(12, 9))

# Линии уровня
ax2.contour(X1, X2, Z, levels=levels, colors='blue', alpha=0.3, linewidths=0.8)
ax2.contour(X1, X2, Z, levels=levels_dense, colors='lightblue', alpha=0.2, linewidths=0.5)

# Траектория Розенброка
ax2.plot(rosen_path[:, 0], rosen_path[:, 1], 'r-', linewidth=2.5, alpha=0.8, label='Траектория')
ax2.scatter(rosen_path[:, 0], rosen_path[:, 1], color='red', s=80, zorder=5, edgecolors='darkred')

# Подписываем точки
for i, (x, y) in enumerate(rosenbrock_points):
    val = f(x, y)
    if i == 0:
        ax2.annotate(f'Начало\n({x}, {y})\nf={val:.1f}', (x, y), xytext=(x-0.2, y-0.15), fontsize=9,
                   bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8),
                   arrowprops=dict(arrowstyle='->', color='gray'))
    else:
        ax2.annotate(f'Ит.{i}\n({x:.2f}, {y:.2f})\nf={val:.1f}', (x, y), xytext=(x+0.07, y+0.07), fontsize=9,
                   bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

# Рисуем направления поиска
for iter_idx, (start_point, directions) in enumerate(zip(rosenbrock_points[:-1], rosenbrock_dirs)):
    start = np.array(start_point)
    for dir_idx, direction in enumerate(directions):
        dir_vec = np.array(direction)
        end = start + dir_vec * 0.6
        ax2.quiver(start[0], start[1], dir_vec[0], dir_vec[1],
                  angles='xy', scale_units='xy', scale=1,
                  color=colors_dir[iter_idx], alpha=0.6, width=0.008,
                  label=f'Направление {iter_idx+1}.{dir_idx+1}' if dir_idx == 0 else "")

ax2.set_xlabel('$x_1$', fontsize=12)
ax2.set_ylabel('$x_2$', fontsize=12)
ax2.set_title('Метод Розенброка: траектория и направления поиска', fontsize=14)
ax2.set_xlim(-0.2, 1.8)
ax2.set_ylim(-0.2, 1.4)
ax2.grid(True, alpha=0.3)
ax2.set_aspect('equal')
ax2.legend(loc='upper left', fontsize=9)

plt.tight_layout()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

def f(x1, x2):
    return (5 - x1)**2 + 10 * (x2 - x1**2)**2

def gradient(x1, x2):
    df_dx1 = -2 * (5 - x1) - 40 * x1 * (x2 - x1**2)
    df_dx2 = 20 * (x2 - x1**2)
    return np.array([df_dx1, df_dx2])

# Параметры Алгоритма 2
alpha_start = 0.01
d = 0.1
x_curr = np.array([0.0, 0.0])
path = [x_curr]
f_vals = [f(*x_curr)]

# Расчет 3 итераций
for _ in range(3):
    s = -gradient(x_curr[0], x_curr[1])
    alpha = alpha_start

    # Шаг 3: Дробление (если условие Армихо не выполнено)
    while f(*(x_curr + alpha * s)) > f(*x_curr) - d * alpha * np.dot(s, s):
        alpha /= 2

    x_curr = x_curr + alpha * s
    path.append(x_curr.copy())
    f_vals.append(f(*x_curr))

path = np.array(path)

# Визуализация
plt.figure(figsize=(12, 8))
x1_range = np.linspace(-0.2, 0.5, 500)
x2_range = np.linspace(-0.1, 0.2, 500)
X1, X2 = np.meshgrid(x1_range, x2_range)
Z = f(X1, X2)

# Фоновые линии уровня
plt.contour(X1, X2, Z, levels=50, cmap='plasma', alpha=0.1)

# Зеленый пунктир через точки
levels = np.sort(np.unique(f_vals))
cs = plt.contour(X1, X2, Z, levels=levels, colors='green', linestyles='dashed', linewidths=1.2)
plt.clabel(cs, inline=True, fontsize=9, fmt='f=%.3f')

# Отрисовка точек и стрелок
for i in range(len(path)):
    lb = f"$x^{i+1} = ({path[i,0]:.4f}, {path[i,1]:.4f})$"
    plt.scatter(path[i, 0], path[i, 1], color='darkmagenta', s=60, label=lb, zorder=5)

    if i < len(path) - 1:
        # Тонкие зеленые стрелочки
        plt.annotate('', xy=(path[i+1, 0], path[i+1, 1]), xytext=(path[i, 0], path[i, 1]),
                     arrowprops=dict(arrowstyle='->', color='green', lw=0.7, shrinkA=0, shrinkB=0))

plt.title('Алгоритм 2: Метод с дроблением шага (Армихо)')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend(title="Координаты точек", loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle=':', alpha=0.4)
plt.tight_layout()
plt.show()
```
